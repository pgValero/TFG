```python
import torch

import numpy as np 
from pathlib import Path, PurePath
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

from skimage.transform import rescale, resize, downscale_local_mean

def img_show(img):
    fig1 = plt.figure(figsize=(6,6), dpi=80)
    # print(img)
    # print(img.shape)
    plt.imshow(img,origin='lower', cmap='YlGnBu', norm=LogNorm())
    plt.colorbar()
    fig1.savefig("Figure_1" + ".pdf", format='pdf',transparent=True)
    plt.show()

def multi_img_show(img_1, img_2, img_3, img_4):
    f, ax = plt.subplots(1, 4)

    # print(img)
    # print(img.shape)
    ax[0].imshow(img_1, origin='lower', cmap='YlGnBu', norm=LogNorm())
    # ax[0].colorbar()
    ax[1].imshow(img_2, origin='lower', cmap='YlGnBu', norm=LogNorm())
    # ax[1].colorbar()
    ax[2].imshow(img_3, origin='lower', cmap='YlGnBu', norm=LogNorm())
    # ax[2].colorbar()
    ax[3].imshow(img_4, origin='lower', cmap='YlGnBu', norm=LogNorm())
    # ax[3].colorbar()
    
    plt.show()
    f.tight_layout()



    f.savefig("Figure_2" + ".pdf", format='pdf',transparent=True)
    
def auto_crop(ori_img = None, dim = 30):
    row_ind = min(ori_img.nonzero()[0][-1] + 2, ori_img.shape[0])
    col_ind = min(ori_img.nonzero()[1][-1] + 2, ori_img.shape[1])
    res = ori_img[row_ind - dim :row_ind, col_ind - dim:col_ind]
    return res


def block_mean(ar, fact):
    assert isinstance(fact, int), type(fact)
    sx, sy = ar.shape
    X, Y = np.ogrid[0:sx, 0:sy]
    regions = sy/fact * (X/fact) + Y/fact
    res = ndimage.mean(ar, labels=regions, index=np.arange(regions.max() + 1))
    res.shape = (sx/fact, sy/fact)
    return res

def load_single_class(class_title = None, downscale = True, autocrop = False):
    path = "sample1"
    class_files = []
    for f in listdir(join(path, class_title)):
        if f.endswith(".npy"):
            class_files.append(f)

    class_img_list = []
    for item in class_files:
        class_img = np.load(join(path,class_title,item))
        if downscale == True:
            class_img = downscale_local_mean(class_img, (20, 16))
        if autocrop == True:
            class_img = auto_crop(ori_img = class_img, dim = 30)

        class_img_list.append(class_img)

    return np.array(class_img_list)



def load_binary(class_1 = "mu+", class_2 = 'e-', downscale = True, autocrop = False):
    # path = "sample1"
    # class_1_files = []
    # for f in listdir(join(path, class_1)):
    #   if f.endswith(".npy"):
    #       class_1_files.append(f)

    # # The dataset of e-
    # class_2_files = []
    # for f in listdir(join(path, class_2)):
    #   if f.endswith(".npy"):
    #       class_2_files.append(f)

    # # Construct the dataset for training
    # class_1_img_list = []
    # class_2_img_list = []

    # for item in class_1_files:
    #   class_1_img = np.load(join(path,class_1,item))
    #   class_1_img_list.append(torch.from_numpy(class_1_img))


    # for item in class_2_files:
    #   class_2_img = np.load(join(path,class_2,item))
    #   class_2_img_list.append(torch.from_numpy(class_2_img))

    # class_1_img_tensor = torch.stack(class_1_img_list)
    # class_2_img_tensor = torch.stack(class_2_img_list)

    class_1_img_list = load_single_class(class_title = class_1, downscale = downscale, autocrop = autocrop)
    class_1_y = np.zeros(len(class_1_img_list))
    class_2_img_list = load_single_class(class_title = class_2, downscale = downscale, autocrop = autocrop)
    class_2_y = np.ones(len(class_2_img_list))

    x_combined = np.concatenate((class_1_img_list, class_2_img_list), axis = 0)
    y_combined = np.concatenate((class_1_y, class_2_y), axis = 0)

    permutation = np.random.permutation(len(x_combined))
    x_combined = x_combined[permutation]
    y_combined = y_combined[permutation]

    x_combined = torch.from_numpy(x_combined)
    y_combined = torch.from_numpy(y_combined).type(torch.LongTensor)


    return x_combined, y_combined
```

```python
from tensorflow import keras
```

```python
x, y = load_binary(class_1 = "mu+", class_2 = 'e-', downscale = True)

x_np = x.numpy()
y_np = y.numpy()

print(x_np.shape)
print(y_np.shape)
```

```python
model = keras.Sequential([
    keras.layers.Conv2D(32, (5, 5), activation='relu',
                        kernel_initializer='he_uniform', padding='same',
                        input_shape=(1, 1, 30, 30), data_format='channels_first'),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu', kernel_initializer='he_uniform'),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(1, activation='softmax'),
])
print(model.output_shape)
model.summary()
```

```python
model.compile(optimizer=keras.optimizers.SGD(lr=0.001, momentum=0.9),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_np, y_np, epochs=8)
```
