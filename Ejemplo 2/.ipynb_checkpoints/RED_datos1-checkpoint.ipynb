```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
dataset1 = pd.read_csv('dataI.csv',sep=';')
X=dataset1.iloc[:,3:-1]
Y=np.ravel(dataset1.label)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=23)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# Define the scaler 
scaler = StandardScaler().fit(X_train)

# Scale the train and test sets
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(200,)),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
print(model.output_shape)
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(#learning_rate=0.8)
                        )

model.compile(loss='binary_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=8, shuffle=True)
```

```python
score = model.evaluate(X_test, Y_test)
print(score)
print(model.metrics_names)
```

```python
resultados = np.around(model.predict(X_test))
```

```python
cm = confusion_matrix(Y_test, resultados, normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap= 'Blues')
```

```python
cm = confusion_matrix(Y_test, resultados, normalize=None)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap= 'Reds')
```

```python
model.save('modelo_datos1.tf',
    overwrite=True,
    include_optimizer=False,
    save_format='tf',
    signatures=None,
    options=None,
    save_traces=True)
```
